```python
import torch

tensor = torch.ones(2, 3)
"""
print(f"First row: {tensor[0]}")
print(f"First column: {tensor[:, 0]}")
print(f"Last column: {tensor[..., -1]}")
"""
tensor[:,1] = 0
print(tensor)

```

两种矩阵乘法

```python
y1 = tensor@tensor.T # 矩阵乘法
print(y1)
z1 = tensor*tensor #对应位置元素相乘
print(z1)
```

```python
x = torch.randn(3, 4)
v = x.view(2, 6)
r = x.reshape(2, 6)
print(x)
print(v)
print(r)
```

```python
v[0, 0] = 999
print(x)
```

view() 遇到不连续，会直接报错！
v = x_t.view(2, 6)  # 运行这行会报错: RuntimeError<br>
但是reshape在内存不连续的时候可以复制一份，克服内存不连续的问题
还可以直接view(-1)将矩阵转换为一维向量，方便查看

```python
x = torch.randn(3, 4)
x_t = x.t()  # 转置操作后，内存变得不连续
print(f"x_t 是否连续: {x_t.is_contiguous()}")  # False
# reshape() 遇到不连续，不会报错，它会自动复制一份新内存来处理
r = x_t.reshape(2, 6)
```

广播机制实现了时两个或两个以上数组的运算，即使这些数组的shape不是完全相同的，只需要满足如下任意一个条件即可。

如果两个数组的后缘维度（trailing dimension，即从末尾开始算起的维度）的轴长度相符，
或

如果两个数组的后缘维度（trailing dimension，即从末尾开始算起的维度）的轴长度其中的一方的长度为1。
